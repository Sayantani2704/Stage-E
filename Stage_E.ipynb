```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
url1 = 'https://investors.coca-colacompany.com/stock-information/year-end-market-values'
req1 = requests.get(url1)
soup1 = BeautifulSoup(req1.text, 'lxml')
soup1.find('table', class_="content-table")
```

```python
data = {'Year':['2009','2010','2011','2012','2013','2014','2015', '2016', '2017', '2018', '2019'], 'Price of Stock (in $)':[57, 65.77, 69.97, 36.25, 41.31, 42.22, 42.96, 41.46, 45.88, 47.35, 55.35]}
data = pd.DataFrame(data)
data
```

```python
plt.plot(data['Year'], data['Price of Stock (in $)'], color='red')
plt.xlabel('Year')
plt.ylabel('Price of Stock (in $)')
plt.title('Coca Cola Company Price of Stock Over Years')
```
